### Data preprocessing

#### Checking null values

```python
df.isnull()
```

```python
df.isnull().sum()
```

- drop column 'wip' as it contains high number of null values

```python
df=df.drop(columns=['wip'])
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.info()
```

#### handling duplicates

```python
#no need to change as there are no duplicates
df.duplicated().sum()
```

#### handling date column

```python
df['date'].describe()
```

```python
#converting date column to datetime format
df['date']=pd.to_datetime(df['date'])
```

```python
#adding new column of month
df['month']=df['date'].dt.month
```

```python
#now no need of date column so remove it
df=df.drop(columns=['date'])
```

```python
df.sample(10)
```

#### handling department column

```python
df['department'].value_counts()
```

```python
#as finishing split in two (maybe due to spaces) hence merge them
df['department']=df['department'].apply(lambda x:'finishing' if x.replace(" ","")=='finishing' else 'sweing')
```

```python
df['department'].value_counts()
```

### Creating Train-Test split

```python
x=df.drop(['actual_productivity'],axis=1);
y=df['actual_productivity']
```

```python
X=x.to_numpy()
```

```python
X
```

```python
#splitting data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
plt.plot(x_train,y_train,'o')
plt.show()
```
